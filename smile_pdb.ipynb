```python
from rdkit import Chem
from rdkit.Chem import AllChem
import os

# SMILES 字符串
test_smile = ['COc1cc2c(Oc3ccc(NC(=O)C4(C(=O)Nc5ccc(F)cc5)CC4)cc3F)ccnc2cc1OCCCN1CCOCC1']

# 转换 SMILES 为 RDKit 分子对象
mols = [Chem.MolFromSmiles(smile) for smile in test_smile]

# 添加氢原子
for i, mol in enumerate(mols):
    mol = Chem.AddHs(mol)
    # 生成三维坐标
    AllChem.EmbedMolecule(mol)
    AllChem.UFFOptimizeMolecule(mol)

    # 保存为 PDB 文件
    pdb_filename = f"6qp5_molecule_{i}.pdb"
    with open(pdb_filename, 'w') as f:
        f.write(Chem.MolToPDBBlock(mol))

    print(f"Saved {pdb_filename}")

# 进入 PyMOL 的命令行（如果您在 Jupyter Notebook 中）
print("You can now load the PDB file in PyMOL.")
```

```python
import matplotlib.pyplot as plt

# 输入数据
indices = [45, 47, 78, 79, 85, 87, 88, 125, 126, 127, 128, 131, 142, 144, 
           145, 146, 147, 148, 150, 151, 154, 171, 172, 174, 176, 261, 262, 
           265, 288, 291, 332, 346, 347, 348, 349, 350, 351, 352]
pocket_seq = "THHIGPHSEKESFHANNIHRKFGSHSTDPEKRHLIDNC"

# 校验长度匹配
assert len(pocket_seq) == len(indices), "序列长度与索引数量不匹配"

# 可视化参数
plt.figure(figsize=(20, 3))
ax = plt.gca()
ax.set_ylim(-0.5, 1.5)
ax.set_xlim(-1, len(pocket_seq) + 1)
ax.axis('off')

# 绘制序列字符及索引
for i, (res, idx) in enumerate(zip(pocket_seq, indices)):
    # 残基字符（上层）
    plt.text(
        x=i, y=1, 
        s=res, 
        ha='center', va='center',
        fontsize=14, fontweight='bold', 
        color='black'
    )
    # 索引数值（下层）
    plt.text(
        x=i, y=0, 
        s=str(idx),
        ha='center', va='center',
        fontsize=10, 
        color='black'
    )

plt.title("Pocket Residues with Original Sequence Indices", fontsize=16, pad=20)
#plt.savefig("pocket_seq_indices.png", dpi=300, bbox_inches='tight')
plt.show()
```
